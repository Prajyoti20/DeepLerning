### **step4-Performs Preprocessing on x**

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),('scalar',StandardScaler())]).set_output(transform='pandas')
```

```python
num_pipe
```

```python
pre=num_pipe.fit_transform(x)
```

```python
pre
```

### **step-5:Spliting of data**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(pre,y,test_size=0.2,random_state=23)
```
