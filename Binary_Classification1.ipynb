###  **Step 3 - Seperate x and y (booking_status) **
### id is Statasticaly insignificant

```python
x=df.drop(columns=["id","booking_status"])
y=df[['booking_status']]
```

### **step4-Performs Preprocessing on x**

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),('scalar',StandardScaler())]).set_output(transform='pandas')
```

```python
num_pipe
```

```python
pre=num_pipe.fit_transform(x)
```

```python
pre
```

### **step-5:Spliting of data**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(pre,y,test_size=0.2,random_state=23)
```

### **Building a Neural Network**

```python
from keras.models import Sequential
from keras.layers import  Input,Dense,Dropout
```

```python
#Initialize the model
model= Sequential()
#Input layer 1
model.add(Input(shape=(x_train.shape[1], )))

#Hidden layer 1
model.add(Dense(64, activation='relu' ))

# add dropout layer 1
model.add(Dropout(0.4))

#Hidden layer 2
model.add(Dense(32, activation='relu' ))

# add dropout layer 2
model.add(Dropout(0.3))

#output layer
model.add(Dense(1, activation='sigmoid' ))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
hist = model.fit(x_train, y_train, epochs=60, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(hist.history['loss'],label = 'Training Loss')
plt.plot(hist.history['val_loss'],label = 'Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Binary cross entropy Loss')
plt.title('Learning curve')
plt.legend()
plt.show()
```

```python
model.evaluate(x_train,y_train)
```

```python
model.evaluate(x_test,y_test)
```

### **Get predictions on train and test**
